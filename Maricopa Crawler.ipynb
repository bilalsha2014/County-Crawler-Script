```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Define the path of the CSV file
csv_file_path = 'test.csv'

# Set up the Selenium driver
driver = webdriver.Chrome()

# Open the CSV file and loop through each row
with open(csv_file_path, 'r', newline='') as csv_read_file:
    reader = csv.reader(csv_read_file)
    next(reader) # skip the header row
    
    # Define the headers for the output CSV file
    headers = ['url', 'data']
    
    # Open the output CSV file and write the headers to the first row
    with open('output.csv', 'w', newline='') as csv_write_file:
        writer = csv.writer(csv_write_file)
        writer.writerow(headers)
        
        for row in reader:
            url = row[0] # extract the URL from the current row
            
            # Load the URL and wait for the elements to become visible
            driver.get(url)
            wait = WebDriverWait(driver, 10)
            elements = wait.until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, 'div#parcel-content')))
            
            # Loop through each element and extract the data
            for element in elements:
                # Get the text of the element
                text = element.text
                
                # Write the URL and data to the output CSV file
                writer.writerow([url, text])
            
# Close the driver
driver.quit()
```
